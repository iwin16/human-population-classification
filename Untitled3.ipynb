```python
sign = [sign0,sign1,sign2_1,sign3,sign4]

# p1 =  pd.DataFrame(columns = list(test.columns))

signdict1 = {}

for p in range(9,63):
    signlist=[]
    
    for item in sign: 
        cout = 0
        
        for i in range(len(item)):
            x = item.iloc[i,0].split("_") 
            temp =test.loc[(test[1])==int(x[1])]
            
            if temp.shape[0] > 0:
                
            
            
                if ((temp.iloc[0,0] == int(x[0])) &  (temp.iloc[0,1] == int(x[1])) & (temp.iloc[0,3] == x[2])
                    & (temp.iloc[0,4] == x[3]) & (temp.iloc[0,p] != '0|0')):
                    cout += 1    
        signlist.append(cout)
    signdict1[p]=signlist
    print(p)
```
